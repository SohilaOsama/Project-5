```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import missingno as msno
# Step 1: Data Understanding and Fast View
# Load the sales data from the CSV file
sales_data = pd.read_csv('sales_data.csv')

# Display the first few rows of the dataset
sales_data.head()
```

```python
sales_data.drop(['Order ID'], axis = 1, inplace = True) 
```

```python
# Step 2: Data Preprocessing
# Check for null values using Missingno library
msno.matrix(sales_data)
plt.show()

```

```python
# Handle null values
sales_data['Product'].fillna(value='Not available', inplace=True)
sales_data['Order Date'].fillna(value='Not available', inplace=True)
sales_data['Category'].fillna(value='Not available', inplace=True)
sales_data['Price Each'].fillna(value=0, inplace=True)
```

```python
sales_data.isnull().sum()
```

```python
# Remove duplicate rows
sales_data.drop_duplicates(inplace=True)
```

```python
# Check for duplicated records
duplicated_records = sales_data.duplicated().sum()
duplicated_records
```

```python
# Convert data types if necessary
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data['Category'] = sales_data['Category'].astype('str') 
```

```python
sales_data.info()
```

```python
# Step 3: Exploratory Data Analysis (EDA)
# Calculate summary statistics
summary_stats = sales_data.describe()
print(summary_stats)
```

```python
import numpy as np
#because in Order ID=3720 outliers and Product_ean =3458 outliers and Quantity Ordered=1154 outliers they represent big number 
#of data for this reason we will treat it with capping

# Step 2: Data Preprocessing - Treat Outliers with Capping

# Identify the columns with outliers
outlier_columns = [ 'Product_ean', 'Quantity Ordered']

# Calculate the upper and lower bounds for each column
for column in outlier_columns:
    lower_bound = sales_data[column].quantile(0.01)
    upper_bound = sales_data[column].quantile(0.99)
    
    # Replace outliers with the respective bounds
    sales_data[column] = np.where(sales_data[column] < lower_bound, lower_bound, sales_data[column])
    sales_data[column] = np.where(sales_data[column] > upper_bound, upper_bound, sales_data[column])

# Step 3: Exploratory Data Analysis (EDA)
# Calculate summary statistics
summary_stats = sales_data.describe()
print(summary_stats)
```

```python
# Specify the columns you want to remove outliers from
columns_to_filter = ["Price Each","turnover", "margin"]

# Define the outlier boundaries and treat the outliers
for column in columns_to_filter:
    q1 = sales_data[column].quantile(0.25)
    q3 = sales_data[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Treat the outliers by replacing them with the column mean
    sales_data[column] = sales_data[column].apply(lambda x: sales_data[column].mean() if x < lower_bound or x > upper_bound else x)

# Save the treated data to a new CSV file
sales_data.to_csv("filtered_file.csv", index=True)
```

```python
# Specify the columns you want to remove outliers from
columns_to_filterd = ["Price Each"]

# Define the outlier boundaries and treat the outliers
for column in columns_to_filterd:
    q1 = sales_data[column].quantile(0.25)
    q3 = sales_data[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Treat the outliers by replacing them with the column median
    sales_data[column] = sales_data[column].apply(lambda x: sales_data[column].median() if x < lower_bound or x > upper_bound else x)

# Save the treated data to a new CSV file
sales_data.to_csv("filtered_file.csv", index=True)
```

```python
# Specify the column you want to treat outliers for
column_to_treat = "Price Each"

# Calculate the Z-scores for the column
z_scores = np.abs((sales_data[column_to_treat] - sales_data[column_to_treat].mean()) / sales_data[column_to_treat].std())

# Define the Z-score threshold for outlier detection
z_score_threshold = 3

# Filter out rows with outliers
df = sales_data[z_scores <= z_score_threshold]

# Save the filtered data to a new CSV file
df.to_csv("filtered_file.csv", index=True)
```

```python
# Specify the column you want to remove outliers from
column_to_filter = "Price Each"

# Calculate the Z-scores for the column
z_scores = np.abs((df[column_to_filter] - df[column_to_filter].mean()) / df[column_to_filter].std())

# Define the Z-score threshold for outlier detection
z_score_threshold = 3

# Filter out rows with outliers
df = df[z_scores <= z_score_threshold]

# Save the filtered data to a new CSV file
df.to_csv("filtered_file.csv", index=False)
```

```python
# Select numerical columns
outlier_columns= df.select_dtypes(include='number').columns
# Create a dictionary to store the number of outliers for each column
outlier_counts = {}

# Calculate the upper and lower bounds for each column
for column in outlier_columns:
    lower_bound = df[column].quantile(0.01)
    upper_bound = df[column].quantile(0.99)
    
    # Find outliers in the column
    outliers = df[column][(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    # Count the number of outliers
    outlier_count = len(outliers)
    
    # Store the number of outliers in the dictionary
    outlier_counts[column] = outlier_count

# Print the table with the number of outliers in each column
print("Number of Outliers:")
for column, count in outlier_counts.items():
    print(f"{column}: {count}")
```

```python
# Question 5: What is the monthly trend of sales turnover?
sales_data['Month'] = sales_data['Order Date'].dt.month
monthly_sales = sales_data.groupby('Month')['turnover'].sum().reset_index()
sns.lineplot(x='Month', y='turnover', data=monthly_sales)
plt.title('Monthly Sales Turnover Trend')
plt.show()
```

```python
# Question 6: What are the top-selling products based on quantity ordered?
top_products = sales_data.groupby('Product')['Quantity Ordered'].sum().reset_index().sort_values('Quantity Ordered', ascending=False)
sns.barplot(x='Quantity Ordered', y='Product', data=top_products.head(10))
plt.title('Top Selling Products')
plt.show()
```

```python
# Question 8: What is the distribution of margins for each product category?
sns.boxplot(x='Category', y='margin', data=df)
plt.title('Distribution of Margins by Category')
plt.show()
```

```python
# Let's extract the city name and create a new column called 'City' using regular expressions
sales_data['City'] = sales_data['Purchase Address'].apply(lambda x: re.findall(r',\s([\w\s]+)\s\d+', x)[0])

# Display the updated dataset with the 'City' column
sales_data.head()
```

```python
city_sales = sales_data.groupby('City')['turnover'].sum().reset_index()
sns.barplot(x='City', y='turnover', data=city_sales)
plt.title('Sales Distribution by City')
plt.xticks(rotation=45)
plt.show()
```
